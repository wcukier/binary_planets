```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def plot_square(min_x, min_y, n, color="black"):
    xs = np.linspace(min_x, min_x + 0.75, 10)
    ys = np.linspace(min_y, min_y + 0.75, 10)

    xx, yy = np.meshgrid(xs, np.flip(ys))

    xx = xx.flatten()
    yy = yy.flatten()

    plt.scatter(xx[:n], yy[:n], color=color)
    # plt.xlim(-0.1,1.1)
    # plt.ylim(-0.1,1.1)

def plot_n(n, color="black"):
    min_x = 0
    min_y = 0
    n_y = 0
    floor_x = 0
    while(n > 0):
        while(n_y < 5):
            while(min_x < 5+floor_x):
                if n >0 : plot_square(min_x, min_y, n, color=color)
                n -= 100
                min_x += 1
            min_x = floor_x
            min_y -= 1
            n_y += 1
        n_y = 0
        min_y = 0
        floor_x +=5.25
        min_x = floor_x
```

```python
plt.figure(figsize=(20,20))
plot_n(4046, color="green")
# plot_n(3069, color="blue")
# plot_n(49, color="red")

plt.gca().set_aspect('equal')
plt.axis('off')
plt.savefig("figs/known_sys_infographic.png", transparent=True)
```

```python
plt.figure(figsize=(20,20))
plot_n(4046, color="green")
plot_n(3069, color="blue")
# plot_n(49, color="red")

plt.gca().set_aspect('equal')
plt.axis('off')
plt.savefig("figs/transit_sys_infographic.png", transparent=True)
```

```python
plt.figure(figsize=(20,20))
plot_n(4046, color="green")
plot_n(3069, color="blue")
plot_n(49, color="red")

plt.gca().set_aspect('equal')
plt.axis('off')
plt.savefig("figs/valid_sys_infographic.png", transparent=True)
```
